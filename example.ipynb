# Example
In this example notebook we show how to simulate the solar system. 

### Step 1: Importing the Module
First we import the module. 

```python
# import the module
from main import Body, SolarSystem
```

### Step 2: Defining the Bodies
We set a starting date and instantiate the bodies in our solar system. Here we simulate the inner solar system. Note that the body name should be given as a string and with a capital. 

```python
# define start date, format YYYY-MM-DD
start_date = '2023-05-16'

# instantiate sun and planets
sun = Body('Sun')

# instantiate planets inner solar system
mercury = Body('Mercury')
venus = Body('Venus')
earth = Body('Earth')
mars = Body('Mars')
```

### Step 3: Constructing and Evolving the Solar System
Now we construct the solar system. We define the timespan over which we simulate the solar system, the timestep and the starting date. Then we add the bodies defined above to the solar system. Finally, we integrate the system over the defined timespan. For the `integrate` method to work, we need at least the sun and one planet in the solar system. 

```python
# instantiate solar system
ss = SolarSystem(earth.T,1.,start_date)

# add sun
ss.add_body(sun)

# add inner planets
ss.add_body(mercury)
ss.add_body(venus)
ss.add_body(earth)
ss.add_body(mars)

# integrate system, using Runge-Kutta by default
ss.integrate()
```

### Step 4: Animation
Finally, generate the animation via the method `animate(size, fps, file_name)`. We define the size of the canvas via the first argument. The whole canvas will have dimensions `2*size x 2*size`. The `size` should be given in astronomical units (AU). The sun will always be place at the center of the canvas. The second argument `fps` sets the frame-per-second of the animation. Finally we set the `file_name`, excluding the `.mp4` extension. 

```python
# make animation movie in mp4 format
ss.animate(2,45, 'inner_solar_system')
```
